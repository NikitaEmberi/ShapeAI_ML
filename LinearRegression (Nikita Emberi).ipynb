```python
import numpy as np
import pandas as pd
import sklearn
```

```python
from sklearn.datasets import load_boston
df = load_boston()
```

```python
df.keys() #Returns all the keys of the dataset dictionary
```

```python
print(df.DESCR) #Description of dataset
```

```python
print(df.feature_names) #Columns
```

```python
print(df.target) #Info about the dataset (Cost of all houses)
```

```python
print(df.data) #Info about the dataset
```

```python
#Convert dictionay to tabular data
boston = pd.DataFrame(df.data,columns = df.feature_names)
boston.head()
#For first 20 rows
#boston.head(20)
```

```python
#Adding a  new column of target values to boston dataframe (cost)
boston['TARGET_VALUE'] = df.target
boston.head()
```

```python
#CHECK IF THE DATASET CONTAINS ANY NULL VALUES OR NOT
boston.isnull()
```

```python
#isnull returns True or False for each of the cell in the dataframe, but we can't go
#exploring all the cells to look for True values if any, so we use sum() function to
#count all the cells with True value(ie Null cells)
boston.isnull().sum()
```

```python
#We never train the model on all the data that we have, we always make sure to atleast have a
#test dataset, which is different from the training dataset

from sklearn.model_selection import train_test_split

X = boston.drop('TARGET_VALUE',axis=1)
#axis = 0 means drop row
Y = boston['TARGET_VALUE']

#X to Predict Y
#X will contain all the columns except Target_value column bcoz we will be using X to predict Y
#Y will have Target_value column (Target value you want to achieve)

#creating tuple
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.15,random_state=5)
#test size : 15% in data points in test and rest 85% in train data points which are selected randomly
#which is specified by random_state
#If you set random_state to a particular number it will choose same numbers everytime you run the code

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#Importing linear date regression model and train it on training dataset
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
#mean_squared_error is basically just to know how much error in the entire predicted value from what we expected from it
```

```python
##Fitting model on the training dataset

lin_model = LinearRegression()
lin_model.fit(X_train,Y_train)
```

```python
#Prediction

y_train_predict = lin_model.predict(X_train)
####Two tests
#First test: on data that we have right now
rmse = (np.sqrt(mean_squared_error(Y_train,y_train_predict)))

print("The model performance for training set")
print(f"RMSE is {rmse}")
print("\n")

#Second test: on testing data
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test,y_test_predict)))
print("The model performance for testing set")
print(f"RMSE is {rmse}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.scatter(Y_test,y_test_predict)
plt.plot([min(y_test_predict),max(y_test_predict)],[min(y_test_predict),max(y_test_predict)])
plt.xlabel('Actual')
plt.ylabel('Predicted')
```
